```python
# make a bin for the Blood Pressure. its Low, High Medium
bins=np.linspace(min(df['Blood Pressure']),max(df['Blood Pressure']),4)
groupnames=['Low', 'Medium', 'High']
df['Blood-binned']= pd.cut(df['Blood Pressure'],bins,labels=groupnames,include_lowest=True)
```

```python
df.head(5)
```

```python
print('Blood pressure Data')
print('the number that is Medium',len(df[df['Blood-binned']=='Medium']))
print('the number that is High' ,len(df[df['Blood-binned']=='High']))
print('the number that us Low', len(df[df['Blood-binned']=='Low']))
```
